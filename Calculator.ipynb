```python
import tkinter as tk

class Calculator:
    def __init__(self, master):
        self.master = master
        self.master.title("Simple Calculator")
        self.master.geometry("300x200")

        self.create_widgets()

    def create_widgets(self):
        # Create input fields for numbers
        self.num1_label = tk.Label(self.master, text="Number 1:")
        self.num1_label.grid(row=0, column=0)
        self.num1_entry = tk.Entry(self.master, width=10)
        self.num1_entry.grid(row=0, column=1)

        self.num2_label = tk.Label(self.master, text="Number 2:")
        self.num2_label.grid(row=1, column=0)
        self.num2_entry = tk.Entry(self.master, width=10)
        self.num2_entry.grid(row=1, column=1)

        # Create dropdown menu for operation
        self.operation_label = tk.Label(self.master, text="Operation:")
        self.operation_label.grid(row=2, column=0)
        self.operation = tk.StringVar(self.master)
        self.operation.set("+")  # default value
        self.operation_menu = tk.OptionMenu(self.master, self.operation, "+", "-", "*", "/")
        self.operation_menu.grid(row=2, column=1)

        # Create button to perform calculation
        self.calculate_button = tk.Button(self.master, text="Calculate", command=self.calculate)
        self.calculate_button.grid(row=3, column=0, columnspan=2)

        # Create label to display result
        self.result_label = tk.Label(self.master, text="Result:")
        self.result_label.grid(row=4, column=0)
        self.result_text = tk.Label(self.master, text="")
        self.result_text.grid(row=4, column=1)

    def calculate(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            operation = self.operation.get()

            if operation == "+":
                result = num1 + num2
            elif operation == "-":
                result = num1 - num2
            elif operation == "*":
                result = num1 * num2
            elif operation == "/":
                if num2 != 0:
                    result = num1 / num2
                else:
                    result = "Error: Division by zero"
            else:
                result = "Error: Invalid operation"

            self.result_text.config(text=str(result))
        except ValueError:
            self.result_text.config(text="Error: Invalid input")

root = tk.Tk()
my_calculator = Calculator(root)
root.mainloop()
```
